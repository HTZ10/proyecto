```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential  # Asegúrate de importar esto
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense  # Asegúrate de importar esto
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_generator = train_datagen.flow_from_directory(
    dataset_dir,
    target_size=(150, 150),  # Ajusta el tamaño de la imagen según sea necesario
    batch_size=16,  # Puedes cambiar el tamaño del lote
    class_mode='binary'  # Dado que solo tienes dos clases: sano, enfermo
)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')  # Clasificación binaria
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=10,  # Ajusta este valor según la cantidad de datos aumentados
    epochs=50,
    verbose=1
)
```

```python
from google.colab import files
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np

# Subir imagen
uploaded = files.upload()

# Tomar el nombre del archivo subido
img_path = next(iter(uploaded))
```

```python
# Cargar la imagen
img = load_img(img_path, target_size=(150, 150))

# Convertir la imagen a un array
img_array = img_to_array(img)

# Expandir dimensiones para que sea compatible con el modelo
img_array = np.expand_dims(img_array, axis=0)

# Preprocesar la imagen (normalizar)
img_array = preprocess_input(img_array)

# Hacer la predicción
prediction = model.predict(img_array)

# Mostrar el resultado
if prediction[0] > 0.5:
    print("Es un maiz enfermo")
else:
    print("Es un maiz sano")
```
